```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("fragmentation1.csv", names = "v")
df2 = pd.read_csv("fragmentation2.csv", names = "v")
```

```python
df.head()
```

```python
ax = plt.gca()

df.plot(kind='line',y='v',ax=ax)
plt.xlabel('t')
plt.ylabel('degree of fragmentation')
plt.title("Fragmentation simulation for n = 500")

plt.show()
```

```python
ax = plt.gca()

df2.plot(kind='line',y='v',ax=ax)
plt.xlabel('t')
plt.ylabel('degree of fragmentation')
plt.title("Fragmentation simulation for n = 1000")

plt.show()
```
